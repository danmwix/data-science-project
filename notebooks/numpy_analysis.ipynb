```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Load the cleaned data
# We use '../' because the notebook is inside the 'notebooks' folder 
# and the data is in the 'data' folder
df = pd.read_csv('../data/vgsales_cleaned.csv')

# 2. Convert Global_Sales to a NumPy Array
# NumPy is required for the 'math' part of your assignment
sales_array = df['Global_Sales'].to_numpy()

# 3. Perform NumPy Math
avg_sales = np.mean(sales_array)
max_sales = np.max(sales_array)
std_dev = np.std(sales_array)

print(f"Dataset Statistics:")
print(f"- Total Games: {len(sales_array)}")
print(f"- Average Sales: {avg_sales:.2f}M")
print(f"- Standard Deviation: {std_dev:.2f}M")

# 4. Matplotlib Visualization: Histogram of Global Sales
plt.figure(figsize=(10, 5))
plt.hist(sales_array, bins=50, range=(0, 5), color='purple', edgecolor='black')
plt.title('Distribution of Global Video Game Sales (0-5M Range)')
plt.xlabel('Sales (Millions of Units)')
plt.ylabel('Number of Titles')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig("../figures/histogram.png")  # Save before showing
plt.show()
```
